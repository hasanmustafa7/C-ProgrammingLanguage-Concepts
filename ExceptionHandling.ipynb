<h1>Exception Handling</h1>

<strong style="color:red">Note: Exception class is the parent class of all exceptions that exist in the .Net Framework.</strong>


```polyglot-notebook
public static class Calculator{
    public static int Divide(int numerator, int denomenator)
    {
        return numerator/denomenator;
    }
}
```

```polyglot-notebook
var res = Calculator.Divide(5,0);
Console.WriteLine(res);
```

        TIP
* Start reading stack trace from bottom to top.
* It will help you understand hierarchy of functions that called one another and eventually the last one that threw the exception.

```polyglot-notebook
try{
    var res = Calculator.Divide(5,0);
    Console.WriteLine(res);
}catch(Exception ex){
    Console.WriteLine("Some error occured");
    Console.WriteLine(ex);
    // throw; // send the error back to caller of this code.
}
```

Stack Trace has two important values <br>
Source -> DLL that caused the exception <br>
Message -> Actual error message. <br>

Multi-Catch Exceptions <br>
Most specific ones come on top, followed by generic ones. 

```polyglot-notebook
try{
    var newRes = Calculator.Divide(4,0);
    Console.WriteLine(newRes);
}catch(DivideByZeroException ex)
{
    Console.WriteLine("Division by zero leads to infinity in maths");
}catch(ArithmeticException ex){
    Console.WriteLine("Artihmetic error");
}
catch(Exception ex){
    Console.WriteLine("You messed up. Debug your code");
}
```

```polyglot-notebook
try{
    var newRes = Calculator.Divide(4,0);    // I Dare you to change second argument to 0. :D
    Console.WriteLine(newRes);
    checked{
        int a = int.MaxValue;
        int b = 1;
        int c = a + b;
    }
}catch(DivideByZeroException ex)
{
    Console.WriteLine("Division by zero : You reached Infinity. Come back!");
}catch(OverflowException ex){
    Console.WriteLine("Oops! Overflow Exception has occured");
}
catch(Exception ex){
    Console.WriteLine("You messed up. Debug your code");
}
```
