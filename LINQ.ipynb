There are extension methods to work with LINQ.

```polyglot-notebook
public class Book
{
     public string Title {get;set;}
     public decimal price {get;set;}

     public static List<Book> GetBooks(){
               return new List<Book>(){
               new Book{Title = "Rich Dad Poor Dad", price = 50},
               new Book{Title = "How to win Friends", price = 55},
               new Book{Title = "How to cook?", price = 100},
               new Book{Title = "Make good software fast", price = 200},
               new Book{Title = "Make good software fast", price = 210},
               new Book{Title = "Take your game to next level", price = 75},
               new Book{Title = "How to play football", price = 150},
               new Book{Title = "How to create good work environment", price = 175},
          };
     }
}



```

```polyglot-notebook
var books = Book.GetBooks();
```

```polyglot-notebook
var cheapBooks =  new List<Book>();

foreach(var book in books) // Linq is an alternative to this code block
{
    if(book.price < 100)
    {
        cheapBooks.Add(book);
    }
}

foreach(var book in cheapBooks)
{
    Console.WriteLine(book.Title + "       |    " + book.price);
}
```

```polyglot-notebook
using System.Linq; // necessary for Linq extension methods to work

var cheapbooks = books.Where(b=>b.price < 100);

foreach(var book in cheapbooks)
{
     Console.Write(book.Title);
     Console.Write("    |    ");
     Console.WriteLine(book.price);
}
```

Func<ClassType,bool> is a special delegate that is called Predicate <br>

.OrderBy()          Sorting <br>
.Where()            Filtering <br>
.Select()           Select is used for Projections/Transformations. <br>  
Where().OrderBy().Select()   Chaining Linq extension methods

```polyglot-notebook
var cheapBooksPrice = books.Where(book => book.price < 100)
    .OrderByDescending(book => book.price)
    .Select(book => book.price);

foreach(var price in cheapBooksPrice)
{
    Console.WriteLine(price);
}
```

Up until now you have seen Linq Extension Methods <br>
There is another syntax we can use LINQ statements with. That is LINQ Query Operators
 <br>

Linq Query Operator always starts with FROM and always finish with a SELECT.



```polyglot-notebook
// LINQ Query Operator
var cheaperBooksTitle = from b in books
                        where b.price < 100
                        orderby b.price descending
                        // select b; // not Console.WriteLine() friendly
                        select new {b.Title,b.price}; // Console.WriteLine() friendly

foreach(var title in cheaperBooksTitle)
{
    Console.WriteLine(title); // Prints like javascript object
}
```

```polyglot-notebook
var singleBook = books.Single(b=>b.Title == "How to win Friends"); // Must be only one and only one matching object in the collection. Else it will throw an exception.
Console.WriteLine(singleBook.price);
```

.Single() will throw an exception when there is not a single match found in the collection. <br>
If it is not known whether or not there will be a match, .SingleOrDefault() can be used instead. It returns default value of the type being returned incase no match is found, instead of raising an exception. <br>

So in case of a class type, deault value will be null. <br>
In case of string, it will be null <br>
In case of int, it will be zero (0). <br>

.First() will return the First Match in the collection and will return even if there is a match upnext in the collection. <br>
It will also throw an exception if no match is found.

.FirstOrDefault() will return the default value of the type being iterated over if there is no match found.

.Skip().Take() are used in paging data.

```polyglot-notebook
Console.WriteLine(books.Max(b=>b.price));
Console.WriteLine(books.Min(b=>b.price));
Console.WriteLine(books.Count(b=>b.Title == "Take your game to next level"));

 
```

```polyglot-notebook
var softwareBook = books.LastOrDefault(b => b.Title == "Make good software fast");
Console.WriteLine(new {Title = softwareBook.Title , Price = softwareBook.price});
```

```polyglot-notebook
var averagePrice = books.Average(b=>b.price);
Console.WriteLine(averagePrice);
```
