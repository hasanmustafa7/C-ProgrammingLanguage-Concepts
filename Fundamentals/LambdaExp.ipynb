<h1>Lambda Expression</h1>

<strong>What is a Lambda Expression?</strong>
Lambda Expression is an anonymous funtions
No access modifier <br>
No Name <br>
No return statement <br>

<strong>Why use?</strong>
For convenience and readability.

```polyglot-notebook
// Normal Function
int Square(int number)
{
    return number * number;
}
```

```polyglot-notebook
Console.WriteLine(Square(3));
```

Same can be done with lambda expressions but with less code.

```polyglot-notebook
// Func<int,int> square = Square; // Also Valid but this is not Lambda Expression
Func<int,int> square = number => number * number; // More like javascript arrow function (IN TERMS OF SYNTAX ONLY)
Console.WriteLine(square(4)); 

delegate int Sum(int num, int num2);
Sum plus = (a,b) => a + b;
Console.WriteLine(plus(2,4));
```

```polyglot-notebook
public class Book
{
     public string Title {get;set;}
     public decimal price {get;set;}
}

List<Book> books = new List<Book>(){
     new Book{Title = "Rich Dad Poor Dad", price = 50},
     new Book{Title = "How to win Friends", price = 55},
     new Book{Title = "How to cook?", price = 100},
     new Book{Title = "Make good software fast", price = 200},
     new Book{Title = "Take your game to next level", price = 75},
     new Book{Title = "How to play football", price = 60},
     new Book{Title = "How to create good work environment", price = 45},
     };
```

```polyglot-notebook
// books cheaper than 100 dollar

var cheapBooks = books.Where<Book>(b => b.price < 100).ToList();

foreach(var book in cheapBooks)
{
    Console.WriteLine(book.Title + " has a price : " + book.price);
}
```

```polyglot-notebook
var expensiveBooks = books.Where<Book>(b=>b.price >=100 && b.price < 150).ToList<Book>();

foreach(var book in expensiveBooks)
{
    Console.WriteLine(book.Title + " has price : " + book.price);
}
```

```polyglot-notebook
var veryExpensiveBooks = books.FindAll(b => b.price > 150);
foreach(var book in veryExpensiveBooks)
{
    Console.WriteLine(book.Title + " costs " + book.price + " dollars");
}
```

<strong>What is a Predicate?</strong> <br>
Specific type of delegate that accepts one input parameter and returns boolean.


```polyglot-notebook
bool IsCheaperThan100Dollars(Book book)  // Predicate method
{
    return book.price < 100;
}
```

```polyglot-notebook
var sameCheapBooks = books.FindAll(IsCheaperThan100Dollars); // Dont add parenthesis because we dont intend to call the predicate. We intend to pass its reference to the FindAll() LINQ method.
foreach(var book in sameCheapBooks)
{
    Console.WriteLine(book.Title + " has a price = " + book.price);
}
```
